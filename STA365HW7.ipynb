### 1.

Part 1: Derive $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma,\boldsymbol\beta_0, \Sigma_\beta)$ for a likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$.

\begin{align*}
    p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma,\boldsymbol\beta_0, \Sigma_\beta) &\propto  p(\textbf{y}_{n\times 1}| \textbf{X}_{n\times p}, \boldsymbol \beta, \Sigma) \times p(\boldsymbol \beta_{p \times 1})\\
    &={} (2\pi )^{-n/2}\det(\Sigma)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right) \times (2\pi )^{-p/2}\det(\Sigma_\beta)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\boldsymbol\beta - \boldsymbol\beta_0)^\top \Sigma_\beta^{-1}(\boldsymbol\beta - \boldsymbol\beta_0)\right)\\
    &\propto{} \exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right) \times  \exp \left(-{\frac {1}{2}}(\boldsymbol\beta - \boldsymbol\beta_0)^\top \Sigma_\beta^{-1}(\boldsymbol\beta - \boldsymbol\beta_0)\right)\\
    &\propto{} \exp \left(-{\frac {1}{2}}({\boldsymbol \beta^\top}\mathbf{X^\top}\Sigma^{-1} \mathbf{X\boldsymbol \beta} - 2{\boldsymbol \beta^\top}\mathbf{X^\top}\Sigma^{-1} \mathbf{y}) \right) \times \exp \left(-{\frac {1}{2}}(\boldsymbol\beta^\top \Sigma_\beta^{-1}\boldsymbol\beta - 2 \boldsymbol \beta^\top \Sigma_\beta^{-1} \boldsymbol\beta_0)\right) \\
    &= \exp \left(-{\frac {1}{2}}({\boldsymbol \beta^\top}\left(\mathbf{X^\top}\Sigma^{-1} \mathbf{X} + \Sigma_\beta^{-1} \right)\boldsymbol \beta - 2{\boldsymbol \beta^\top}\left(\mathbf{X^\top}\Sigma^{-1} \mathbf{y} + \Sigma_\beta^{-1} \boldsymbol\beta_0\right))\right) \\
    &\propto \exp \left( -\frac{1}{2} \left( \beta - (\mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma^{-1}_\beta)^{-1}(\mathbf{X}^\top \Sigma^{-1} \mathbf{y} + \Sigma^{-1}_\beta \mathbf{\beta}_0) \right)^\top \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma_\beta^{-1} \right) \left( \beta - (\mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma^{-1}_\beta)^{-1}(\mathbf{X}^\top \Sigma^{-1} \mathbf{y} + \Sigma^{-1}_\beta \mathbf{\beta}_0) \right) \right)\\
    \therefore p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma,\boldsymbol\beta_0, \Sigma_\beta) &= \mathcal{MVN}\left( E[\mathbf{\beta}] = COV[\mathbf{\beta}]^{-1} \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{y} + \Sigma^{-1}_\beta \mathbf{\beta}_0 \right), COV[\mathbf{\beta}] = \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma^{-1}_\beta \right) \right)
\end{align*}

Part 2: Derive $p(\sigma^2 | \mathbf{y}, \mathbf{X}, \alpha^*, \beta^*)$ for a likelihood proportional to a normal and a $\text{Inv-Gamma}(\alpha^*, \beta^*)$ for $\sigma^2$.

\begin{align*}
    p(\sigma^2 | \mathbf{y}, \mathbf{X}, \alpha^*, \beta^*) &\propto p(\mathbf{y}|\mathbf{X}, \mathbf{\beta}, \sigma^2) \times p(\sigma^2)\\
    &\propto (\sigma^2)^{-\frac{1}{2}(n-p)} \exp \left( -\frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\mathbf{\beta})^\top (\mathbf{y}-\mathbf{X}\mathbf{\beta}) \right) \times (\sigma^2)^{-a^*-1} \exp \left( -\frac{\beta^*}{\sigma^2} \right)\\
    &\propto (\sigma^2)^{-(\alpha^*+\frac{n-p}{2}+1)} \exp \left( -\frac{\beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X}\mathbf{\beta})^\top (\mathbf{y}-\mathbf{X}\mathbf{\beta})}{\sigma^2} \right)\\
    \therefore p(\sigma^2 | \mathbf{y}, \mathbf{X}, \alpha^*, \beta^*) &\propto \text{Inv-Gamma}\left( \alpha^*+\frac{n-p}{2}, \beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X}\mathbf{\beta})^\top (\mathbf{y}-\mathbf{X}\mathbf{\beta}) \right)
\end{align*}

### 2.

With a dataset of your choosing, provide Bayesian inference w/ MCMC diagnostics with a non **inverse-gamma** prior for `sigma`, all the while using `pm.Normal('betas', mu=0, sigma=1, shape=p)` and `pm.Normal('y', mu=X@betas, sigma=sigma, observed=y)` rather than `pm.MvNormal` instead of built-in `pm.MvNormal` function.

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
import arviz as az

df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "andrewsundberg/college-basketball-dataset",
  "cbb.csv",
)

df.head()
```

```python
cols = ["EFG_O", "EFG_D", "TOR", "TORD", "ORB", "DRB", "FTR", "FTRD", "ADJOE"]
df = df.dropna()[cols]

y = df["ADJOE"].values  # Basketball teams 'offensive efficiency
X = df.drop(columns=["ADJOE"]).values
n, p = X.shape

with pm.Model() as ADJOE:
    # Priors
    betas = pm.Normal("betas", mu=0, sigma=1, shape=p)
    sigma = pm.HalfNormal("sigma", sigma=1)

    # Likelihood
    y_obs = pm.Normal("y", mu=X@betas, sigma=sigma, observed=y)

    idata = pm.sample(2000, tune=1000, cores=4)

pm.model_to_graphviz(ADJOE)
```

```python
az.summary(idata, hdi_prob=0.95)
```

The interpretations of note here is that the offensive field goal percentage (EFG_O) is a major contributor to the team's ADJOE, and with the $0 \notin \text{HDI}$, we are confident that the positive relationship of 1.754 between EFG_O and ADJOE is prevalent.

Similarly, ball turnover rate (TOR) has a negative relationship with ADJOE for a factor of 1.487, with the rationale for our confidence being akin to EFG_O. This makes sense, because an estimate of the offensive efficiency (points scored per 100 possessions) for a team who turns over often is sure to be low.

```python
az.plot_trace(idata, compact=True)
```

```python
az.plot_energy(idata)
```

### 3.

Let $p(\tau_i)$ be $\require{cancel} \textrm{gamma}\big(\tau_i | \alpha = \frac{\nu}{2}, \overset{\textrm{rate}\xcancel{\textrm{scale}}}{\beta = \frac{\nu}{2}}\big)$ and let $p(y_i|\boldsymbol \beta, \tau,\tau_i)$ be $\mathcal{N}(y_i | \textbf{X} \boldsymbol \beta, \overset{\textrm{precision}}{\tau \times \tau_i})$.

Do the same thing as Q2, but add outliers, find an appropriate prior, use the posterior distributions of the $\tau_i$'s to identify data point "outliers", and use the posterior distributions of the $\sigma_i^{-2} = \tau \times \tau_i$ to create posterior distributions of the **influence** (the diagonals of the $H$ "hat" matrix $X^\top (X^\top D X)^{-1} X$ where $D_{ij}=0$ and $D_{ii} = \sigma^2_i$) and compare and contrast some example "outlier" versus "non outlier" data points.

```python
outlier_indices = np.random.choice(n, size=int(0.05 * n), replace=False)
y[outlier_indices] += np.random.normal(50, 10, size=len(outlier_indices))

with pm.Model() as ADADJOE:
    # Priors
    betas = pm.Normal("betas", mu=0, sigma=1, shape=p)
    nu = pm.Gamma("nu", alpha=2, beta=0.1)
    tau = pm.Gamma("tau", alpha=2, beta=2)
    tau_i = pm.Gamma("tau_i", alpha=nu/2, beta=nu/2, shape=n)
    sigma_i = (tau * tau_i)**(-1)

    # Likelihood
    y_obs = pm.Normal("y", mu=X@betas, sigma=sigma_i, observed=y)

    idata2 = pm.sample(2000, tune=1000, cores=4)

pm.model_to_graphviz(ADADJOE)
```

```python
az.summary(idata2, var_names=["betas", "tau", "nu"], hdi_prob=0.95)
```

```python
az.plot_trace(idata2, var_names=["betas", "tau", "nu"])
```

```python
az.plot_energy(idata2)
```

```python
tau_i_post_means = idata2.posterior["tau_i"].mean(dim=["chain", "draw"])
outlier_indices = tau_i_post_means.argsort().values[:10] # Since we modified our data to have 10 outliers, get 10 smallest tau_i's in our posterior

with ADADJOE:
    post_sigma_sq = (idata2.posterior["tau"] * idata2.posterior["tau_i"])**(-1)
    D_diag = post_sigma_sq.mean(dim=["chain", "draw"]).values

    H_diag = np.diag(X @ np.linalg.inv(X.T @ (D_diag[:, None] * X)) @ X.T)

# Plottiong a histogram of all hat matrix diagonal values
fig, ax = plt.subplots()
ax.hist(H_diag, bins=30, alpha=0.7, label="All Points")
ax.hist(H_diag[outlier_indices], bins=10, alpha=0.7, color="red", label="Outliers")
ax.set_xlabel("Influence (Hat Matrix Diags)")
ax.set_ylabel("Freq")
ax.legend()
plt.show()
```

Here, we can see that the majority of the outliers have smaller influence, which means that in our model, they are unusual but not influential, further meaning the model is already downweighting them effectively (model robustness).
