### Q3.

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```

1. What is this algorithm of a slice sampler actually doing?
- Starting from an initial point $x$, the algorithm iteratively samples a height
$y$ uniformly below $f(x) := \text{Beta}(x|1.5, 3)$ to define a slice $[a,b]$ where $f(x) > y$, where $y \sim \text{Unif}(0, f(x^{(t)}))$. A new candidate $x'$ is sampled uniformly within this interval. If $f(x') > y$, it is accepted; otherwise, the interval is narrowed depending on $x'$ being left or right of $x$, and the process repeats recursively. The intervals are adaptively adjusted until a valid $x'$ is found, and this continues for each step increment of $t$. Similar to what we've seen with Metropolis-Hastings, this method avoids the need for a proposal distribution by directly sampling from regions where the target function is nonzero.

2. How can we use this sampler in place of MH in Gibbs sampling instance?
- Well, instead of proposing a new value and accepting/rejecting as in MH, we would be starting with the current value of the variable, sampling a slice height $y$ beneath the unnormalized full conditional $f(x)$, finding the slice interval $[a, b]$ where $f(x) > y$, and uniformly sampling a new $x$ within this interval. The curve we are sampling beneath in slice sampling is the unnormalized full conditional density of the variable we are updating in each of the Gibbs sampling iteration. The initial value, $x^{(t)}$, is the current value of the variable $x$, and we sample a slice height $y$, find the interval where $f(x) > y$, and sampling a new $x'$ within the interval. This new $x'$ serves as the Gibbs update for that variable, continuing the chain in this MCMC algorithm.
